```python
 from queue import *
```

```python
m = 48 # the number of cores
```

```python
with open("ptf_feat3.txt") as f:
    lines = f.read().rstrip()

trees = lines.split("\n\n") # a list of  strings (each string represents a tree)

assert(len(trees) == m) # one decision tree for each core
```

```python
LEAF_COND = "true"  # condition for leaf node

NO_CHILD = 0 # child for leaf node

LEAF_VAL = -1 # classified value for non-leaf node
```

```python

conds = []
children = []
vals = []

for tree in trees:
    edges = tree.split("\n")  # a list of  strings (each string represents an edge)
    
    n = len(edges)  
    # n = the number of edges
    # n + 1 = the number of nodes
    
    if n == 1: # (no branching at all)
        conds.append([LEAF_COND])
        children.append([NO_CHILD])
        vals.append([edges[0][-1]])  # e.g. get '1' from [ ': 1' ]
        continue
    
    print()
    print(edges)
    print()
    
    cond = ["" for _ in range(n + 1)]
    child = [None for _ in range(n + 1)]
    val = [None for _ in range(n + 1)]
 
    q = Queue()
    
    for e in edges:
        print(e)
        q.put(e)

    for _ in range(n):
        pass
            
    
    # collect each core...
    conds.append(cond)
    children.append(child)
    vals.append(val)



# 2.114625
```
